```python
import os
import json
from tqdm import tqdm

def build_inverted_index(clean_folder):
    inverted = {}

    files = [f for f in os.listdir(clean_folder) if f.endswith(".json")]

    for fname in tqdm(files, desc="Building Inverted Index"):
        doc_id = fname.replace(".json", "")

        with open(os.path.join(clean_folder, fname), "r", encoding="utf-8") as f:
            data = json.load(f)
            tokens = data["tokens"]

        unique_terms = set(tokens)

        for term in unique_terms:
            if term not in inverted:
                inverted[term] = []
            inverted[term].append(doc_id)

    return inverted

inverted_index = build_inverted_index("dataset_clean")

with open("inverted_index.json", "w", encoding="utf-8") as f:
    json.dump(inverted_index, f, indent=2)

print("Selesai, file disimpan dengan nama inverted_index.json")
```
