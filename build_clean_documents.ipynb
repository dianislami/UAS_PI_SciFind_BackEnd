```python
import os
import json

SCRAPE_FOLDER = "dataset_raw"      
CLEAN_FOLDER = "dataset_clean"      
OUTPUT_FILE = "clean_documents.json"

data = []

for fname in os.listdir(SCRAPE_FOLDER):
    if not fname.endswith(".txt"):
        continue

    raw_path = os.path.join(SCRAPE_FOLDER, fname)
    clean_path = os.path.join(CLEAN_FOLDER, fname.replace(".txt", ".json"))

    # cek apakah token file ada
    if not os.path.exists(clean_path):
        print("SKIP (token tidak ditemukan):", clean_path)
        continue

    # === TITLE ===
    title = fname.replace(".txt", "").strip()
    title_clean = title.lower()

    # === DESKRIPSI & POSTER ===
    with open(raw_path, "r", encoding="utf-8") as f:
        raw = f.read().strip()

    lines = raw.split("\n")
    poster = ""
    description = raw

    # kalau baris pertama mengandung link poster
    if lines[0].lower().startswith("poster:"):
        poster = lines[0].split("poster:",1)[1].strip()
        description = "\n".join(lines[1:]).strip()

    # === TOKENS ===
    with open(clean_path, "r", encoding="utf-8") as f:
        tokens = json.load(f)["tokens"]

    # === inject TITLE ke tokens ===
    tokens.append(title_clean)                  
    tokens.extend(title_clean.split())          

    # === Simpan ===
    data.append({
        "title": title,
        "poster": poster,
        "description": description,
        "tokens": tokens
    })

print("Total dokumen berhasil dirakit:", len(data))

with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2, ensure_ascii=False)

print("DONE → clean_documents.json berhasil dibuat!")
```
